```python
import numpy as np
import pandas as pd

# Make dictionary
data_dict = {
    'no.': ['1번','2번','3번','4번'],
    'name': ['김기남', '이건우', '윤혜선', '이정은'],
    'hobby': ['탁구', '축구', '수영', '그림그리기']
}
data_dict
```

```python
# Make a dataframe
data = pd.DataFrame(
    data_dict
)
data
```

```python
# Save .csv file
data.to_csv('output_230913.csv', index=True)
```
