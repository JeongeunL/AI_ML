```python
weather_2 = pd.read_csv('data/weather_by_station.csv', index_col='date', parse_dates=True)
weather_2.head()
```

```python
# 가장 비가 많이 온 분기
weather_2.query('datatype == "PRCP"').groupby(
    ['station', pd.Grouper(freq='Q')]
).sum().unstack().sample(5, random_state=1)
```

```python
## 3. 가장 비가 많이 온 날이 있는 분기
weather_2[weather_2.datatype == "PRCP"]\
      .groupby(level=0).mean()\
      .groupby(pd.Grouper(freq='Q')).sum().value.nlargest()
```

```python
## 3.(2) 분기별 총 강우량
weather_2[weather_2.datatype == "PRCP"]\
      .groupby(level=0).mean()\
      .groupby(pd.Grouper(freq='Q')).sum()
```

```python
## 3.(3) 분기별 총 강우량, 달별 일수
weather_2\
    .query('datatype == "PRCP"')\
    .rename(dict(value='prcp'), axis=1)\
    .groupby(level=0).mean()\
    .assign(
        total_prcp_in_Quarter=lambda x: \
            x.groupby(pd.Grouper(freq='Q')).transform(np.sum),
        total_day_count_in_month=lambda x: \
            x.groupby(pd.Grouper(freq='M')).prcp.transform('count'),
        pct_monthly_prcp=lambda x: \
            x.prcp.div(x.total_prcp_in_Quarter)
    )\
    .nlargest(5, 'pct_monthly_prcp')
```

```python
fb = pd.read_csv('data/fb_2018.csv', index_col='date', parse_dates=True).assign(
    trading_volume=lambda x: pd.cut(x.volume, bins=3, labels=['low', 'med', 'high'])
)
fb.head()
```

```python
# 5. crosstab 연습하기
pd.crosstab(
    index=fb.trading_volume,
    columns=fb.index.month,
    colnames=['month']
    )
```
