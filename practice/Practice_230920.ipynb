```python
import pandas as pd

# 1. Load earthquakes data
df = pd.read_csv('data/earthquakes.csv')
```

```python
df.head()
```

```python
# 2.(1) loc 사용
df.loc[3:10, ['mag', 'magType', 'type']]
```

```python
df.columns
```

```python
# 2.(2) iloc 사용
df.iloc[3:10, [8, 9, 21]]
```

```python
df.place.str.extract(r', (.*$)')[0].sort_values().unique() # place에 들어가 있는 지역 이름 확인
```

```python
# 3. Arizona 지역의 데이터 불러오기
df.loc[
    (df.place.str.contains(r'Arizona$'))
]

# df.loc[df.place.str.contains('Arizona')].filter(['mag', 'magType', 'type'])
```

```python
# 4. mag 값의 상위 10개 값
df.loc[df['mag'].nlargest(10).index, ['mag']]

# (head(10)만 사용하려면 먼저 mag를 크기 기준으로 정렬해야 한다.)
# (하지만, nlargest를 사용하면 사전에 따로 데이터를 정렬할 필요가 없어서 편리하다. )
```

```python
# 5. alert 변수 버리기
del df['alert']
df.columns
```

```python
# 6. mag와 magType 변수를 각각 pop하기
mag = df.pop('mag')
magType = df.pop('magType')
df.columns

# (del과 pop은 다른 개념. pop은 완전히 삭제하는 게 아니라, 데이터프레임에서 꺼낸다는 뜻)
```

```python
df.mag
```

```python
mag
# pop한 후에 df.mag로 실행하면 안 되지만, mag로만 실행하면 나온다.
```

```python
# 7. mag와 magType 변수를 concat하여 새 데이터프레임 만들기
new_df = pd.concat([mag, magType], axis=1)
new_df

# (default는 행으로 concat되기 때문에, axis=1을 넣어줘야 열 방향으로 concat된다.)
```
